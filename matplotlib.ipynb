```python
# VISUALISING A THREE DIMENSIONL FUNCTION
def f(x,y):
    return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
x=np.linspace(0,5,50)
y=np.linspace(0,5,40)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
plt.contour(X,Y,Z)
# WHEN A SINGLE COLOR IS USED,NEGATIVE VALUES ARE REPRESENTED BY DASHED
# LINES AND POSITIVE VALUES BY SOLID LINES
```

```python
plt.plot(X,Y,'.y')
```

```python
plt.contour(X,Y,Z,cmap='RdGy')
```

```python
plt.contour(X,Y,Z,20,cmap='RdGy')
# WE WANT MORE LINES TO BE DRAWN -20 EQUALLY SPACED INTERVALS WITHIN
# THE DATA RANGE
```

```python
# FILLED CONTOUR PLOT
plt.contourf(X,Y,Z,20,cmap='RdGy')
plt.colorbar()

# ONE POTENTIAL ISSUE WITH THIS PLOT IS THAT THE COLOR STEPS ARE
# DISCREET RATHER THAN CONTINOUS WHICH IS NOT ALWAYS WHAT IS DESIRED.
# ALTERNATE FOR THIS IS TO USE PLT.IMSHOW() FUNCTION,WHICH INTERPRETS
# A TWO-DIMENSIONAL GRID OF DATA AS AN IMAGE
```

```python
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')

# POTENTIAL ISSUES WITH IMSHOW()
# 1.DOESN'T ACCEPT AN X AND Y GRID SO U MUST MANUALLY SPECIFY THE 
# EXTENT[XMIN,XMAX,YMIN,YMAX] OF THE IMAGE ON THE PLOT
# 2.PLT.IMSHOW() BY DEFAULT FOLLOWS THE STANDARD IMAGE ARRAY DEFINITON
# WHERE THE ORIGIN IS THE UPPER LEFT,NOT IN THE LOWER LEFT AS IN MOST 
# CONTOUR PLOTS.THIS MUST BE CHANGED WHEN SHOWING GRIDDED DATA
# 3.PLT.IMSHOW() WILL AUTOMATICALLY ADJUST THE AXIS ASPECT RATIO TO
# MATCH THE INPUT DATA,YOU CAN CHANGE THIS BY SETTING ,FOR EG PLT.AXIS
# (ASPECT='IMAGE') TO MAKE X AND Y UNITS MATCH`
```

```python
# SOMETIMES IT CAN BE USEFUL TO COMBINE CONTOUR PLOTS AND IMAGE PLOTS
contours=plt.contour(X,Y,Z,3,colors='black')
plt.clabel(contours,inline=True,fontsize=8)
# plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap='RdGy',alpha=0.5)
# plt.colorbar()
```

```python
# USING PLT.FUNCFORMATTER WHICH ACCEPTS A USER-DEFINED FUNCTION GIVING
# FINE-GRAINED CONTROL OVER THE TICK OUTPUTS
fig,ax=plt.subplots()
x=np.linspace(0,3*np.pi,1000)
ax.plot(x,np.sin(x),lw=3,label='sine')
ax.plot(x,np.cos(x),lw=3,label='cosine')
plt.legend()
ax.axis('equal')


ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))
def format_func(value,ticknumber):
    N=int(np.round(2*value/np.pi))
    if N==0:
        return '0'
    elif N==1:
        return '$\pi/2$'
    elif N==2:
        return '$\pi$'
    elif N%2>0:
        return '${0}\pi/2$'.format(N)
    else:
        return '${0}\pi$'.format(N//2)
    
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))    
    
```

```python
# switching to a stylesheet
# plt.style.use('styleName')
# keep in mind that this will change the style for the rest of the session
# alternatively you can use the style context manager,which sets a style
# temporarily
# with plt.style.context('styleName'):
#     make_a_plot()

# LETS CREATE A FUNCTTION THAT WILL MAKE TWO BASIC TYPES OF PLOTS
def hist_and_lines():
    np.random.seed(0)
    fig,ax=plt.subplots(1,2,figsize=(11,4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a','b','c'],loc='lower left')    
```

```python
# THREE DIMENSIONAL PLOTTING IN MATPLOTLIB
# we enable 3d plots by importing the mplot3d toolkit
from mpl_toolkits import mplot3d
```

```python
# once this module is imported,we can create a 3d axes by passing the
# keyword projection='3d' to any of the normal axes creation routines
ax=plt.axes(projection='3d')
```

```python
# 3D POINTS AND LINES
ax=plt.axes(projection='3d')
z=np.linspace(0,15,1000)
x=np.sin(z)
y=np.cos(z)
ax.plot3D(x,y,z,color='m')
```

```python
# 3d scattered plots
ax=plt.axes(projection='3d')
z=15*np.random.random(100)
x=np.sin(z)+0.1*np.random.randn(100)
y=np.cos(z)+0.1*np.random.randn(100)
ax.scatter3D(x,y,z,c=z,cmap='Greens')

# notice that by default the scatter points have their transparency 
# adjusted to give a sense of depth on the page
```

```python
# 3D CONTOUR PLOTS
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2+Y**2))
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
```

```python
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2+Y**2))
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')

# sometimes the default viewing angle is not optimal,in which case we can 
# use the view_init method to set the elevation and azimuthal angles
# .in the ex we'll use an elevation of 60 degrees (i.e is 60 degrees above
# the x-y plane) and an azimuth of 35 degrees (i.e rotated 35 degrees
# counter clockwise about the z-axis)
ax.view_init(60,35)
```

```python
# WIREFRAMES AND SURFACE PLOTS
# two other types of 3d plots that work on gridded data are wireframes 
# and surface plots
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='black')
```

```python
# a surface plot is like a wireframe plot but each face of the wireframe
# is a filled ploygon
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='viridis')
```

```python
# SURFACE TRIANGULATIONS
# for some applications ,the evenly sampled grids required by the preceed
# ing routines are overly restrictive and inconvenient.In these situations,
# the triangulation based plots can be very useful.What if rather an even
# draw from a cartesian or a polar grid,we have a set of random draws?
theta=2*np.pi*np.random.random(1000)
r=6*np.random.random(1000)
x=np.ravel(r*np.sin(theta))
y=np.ravel(r*np.cos(theta))
z=np.sin(np.sqrt(x**2+y**2))


ax=plt.axes(projection='3d')
ax.scatter3D(x,y,z,c=z,cmap='viridis')
```

```python
# ax.plot.trisurf creates a surface by first finding a set of triangles
# formed between adjacent points
ax=plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='viridis')
```

```python
# VISUALIZATION WITH SEABORN
# matplotlib is not designed for use with pandas DataFrame.
# seaborn provides an API on top of matplotlib thsat offers the same choices
# for plot style and color defaults, defines simple high-level functions
# for common statisticl plot types,and integrates with the functionality
# provided by Pandas DataFrame
rng=np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(rng.randn(500,6),0)
```

```python
# plot with matplotlib defaults
plt.plot(x,y)
plt.legend(['A','B','C','D','E','F'],ncol=2,loc='upper left',frameon=True)
```
